# Pandas Snippet

```python
import pandas as pd
import numpy as np
```

```python
# 데이터 세팅

a1 = [60, 84, 80]
a2 = [77, 62, 95]
a3 = [61, 97, 72]

df_data = [a1, a2, a3]
other_data = [[1,10,100],[2,20,200],[3,30,300]]

cols = ["국어", "영어", "수학"]
rows = ['a','b','c']

df = pd.DataFrame(data=df_data, index=rows, columns=cols)
other = pd.DataFrame(data=other_data,index=rows,columns=cols)
```

```python
# 왼쪽 -> 오른쪽 더하기
display(df.add(other, axis='columns', level=None, fill_value=None))
# 오른쪽 -> 왼쪽 더하기
display(df.radd(other, axis='columns', level=None, fill_value=None))
```

```python
# 왼쪽 -> 오른쪽 빼기
display(df.sub(other, axis='columns', level=None, fill_value=None))
# 오른쪽 -> 왼쪽 빼기
display(df.rsub(other, axis='columns', level=None, fill_value=None))
```

```python
# 왼쪽 -> 오른쪽 곱하기
display(df.mul(other, axis='columns', level=None, fill_value=None))
# 오른쪽 -> 왼쪽 곱하기
display(df.rmul(other, axis='columns', level=None, fill_value=None))
```

```python
# 왼쪽 -> 오른쪽 나눗셈
display(df.div(other, axis='columns', level=None, fill_value=None))
# 오른쪽 -> 왼쪽 나눗셈
display(df.rdiv(other, axis='columns', level=None, fill_value=None))
```

```python
# 왼쪽 -> 오른쪽 나머지
display(df.mod(other, axis='columns', level=None, fill_value=None))
# 오른쪽 -> 왼쪽 나머지
display(df.rmod(other, axis='columns', level=None, fill_value=None))
```

```python
# 왼쪽 -> 오른쪽 거듭제곱
display(df.pow(other, axis='columns', level=None, fill_value=None))
# 오른쪽 -> 왼쪽 거듭제곱
display(df.rpow(other, axis='columns', level=None, fill_value=None))
```
